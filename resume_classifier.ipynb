```python
import os
from pdf2docx import Converter

def convert_pdf_to_docx(pdf_folder, docx_folder):
    # Create the output directory if it doesn't exist
    if not os.path.exists(docx_folder):
        os.makedirs(docx_folder)

    # Iterate over all files in the PDF folder
    for filename in os.listdir(pdf_folder):
        if filename.endswith('.pdf'):
            # Construct paths for PDF and DOCX files
            pdf_path = os.path.join(pdf_folder, filename)
            docx_path = os.path.join(docx_folder, os.path.splitext(filename)[0] + '.docx')

            # Convert PDF to DOCX

            cv = Converter(pdf_path)
            cv.convert(docx_path, start=0, end=None)
            cv.close()

# Set the paths for the PDF folder and the output DOCX folder
pdf_folder = '/content/drive/MyDrive/Resumes'
docx_folder = '/content/drive/MyDrive/Resume Classification/MDE_docx'

# Convert PDFs to DOCXs
convert_pdf_to_docx(pdf_folder, docx_folder)
```

```python

# try to extract all folder one time
import os

file_name = []
category1= []
base_dir = './Data Engineer'
for root, dirs, files in os.walk(base_dir, topdown=False):
    for name in files:
        if name.endswith(".docx"):
            file_path = os.path.join(root, name)
            file_name.append(file_path)
```

```python
file_path1 = []
category1  = []
directory1 = '/content/drive/MyDrive/Resumes/Data Engineer'
for i in os.listdir(directory1):
    if i.endswith('.docx'):
        os.path.join(directory1, i)
        file_path1.append((textract.process(os.path.join(directory1, i))).decode('utf-8'))
        category1.append('Data Engineer')
```

```python
data1 = pd.DataFrame(data = file_path1 , columns = ['Raw_Details'])
data1['Category1'] = category1
data1
```

```python

file_path2 = []
category2  = []
directory2 = '/content/drive/MyDrive/Resumes/Data Scientist'
for i in os.listdir(directory2):
    if i.endswith('.docx'):
        os.path.join(directory2, i)
        file_path2.append((textract.process(os.path.join(directory2, i))).decode('utf-8'))
        category2.append('Data Scientist')
```

```python
data2 = pd.DataFrame(data = file_path2 , columns = ['Raw_Details'])
data2['Category2'] = category2
data2
```

```python

file_path4 = []
category4  = []
directory4 = '/content/drive/MyDrive/Resumes/Java Developer'
for i in os.listdir(directory4):
    if i.endswith('.docx'):
        os.path.join(directory4, i)
        file_path4.append((textract.process(os.path.join(directory4, i))).decode('utf-8'))
        category4.append('Java Developer')
```

```python
data4 = pd.DataFrame(data = file_path4 , columns = ['Raw_Details'])
data4['Category4'] = category4
data4
```

```python
resume_data = pd.concat([data1, data2,data4], ignore_index=True)
resume_data
```
